 Question 1: Traditional Machine Learning Model

In this question, traditional machine learning classifiers are used to distinguish
confirmed exoplanets from false positives using features extracted from the Kepler
space telescope dataset. The task is formulated as a binary classification problem,
where confirmed planets are labelled as 1 and non-confirmed detections are labelled as 0.

```python
# imports
import sys, os, pandas as pd
from importlib import reload
import kagglehub
!rm -rf /content/up2115556-machine-learning-and-neural-network-coursework-2 # ensures a clean cell (this was a massive issue over and over)
!git clone https://github.com/up2115556/up2115556-machine-learning-and-neural-network-coursework-2.git # adds the repository
sys.path.insert(0, "/content/up2115556-machine-learning-and-neural-network-coursework-2")
# takes my helper functions from the git
from helpers.functions import prepare_kepler_data, make_train_test_split
import helpers.functions as funcs
reload(funcs)
prepare_kepler_data = funcs.prepare_kepler_data
make_train_test_split = funcs.make_train_test_split
#downloads kepler
path = kagglehub.dataset_download("nasa/kepler-exoplanet-search-results")
df = pd.read_csv(f"{path}/cumulative.csv")
X, y = prepare_kepler_data(df)
#prepares features and scales them
X_train_scaled, X_test_scaled, y_train, y_test, scaler = make_train_test_split(X, y)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

Logistic Regression

Logistic Regression is used as a baseline traditional machine learning model.
It is a linear classifier that estimates the probability of a detection being
a confirmed exoplanet based on the input features.

```python
#trains and runs the logitstic regression model
log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train_scaled, y_train)
y_pred_lr = log_reg.predict(X_test_scaled)
print("logistic regression")
print("confusion matrix:")
print(confusion_matrix(y_test, y_pred_lr)) # this matrix is predicted not predicted on top vs actual not vs actual down the side
print("\nclassification report:")
print(classification_report(y_test, y_pred_lr))
print("accuracy:", accuracy_score(y_test, y_pred_lr))
```

The Logistic Regression model achieves an accuracy of approximately 91%.
It demonstrates high recall for confirmed exoplanets, indicating that most real
planets are correctly identified. However, the lower precision suggests that some false positives are incorrectly classified as planets.


Random Forest

A Random Forest classifier is used to model more complex, non-linear relationships between the features. It combines multiple decision trees to improve speed.


```python
#trains and runs the random forest
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
print("\n\nrandom forest")
print("confusion matrix:")
print(confusion_matrix(y_test, y_pred_rf)) # this matrix is predicted not predicted on top vs actual not vs actual down the side
print("\nclassification report:")
print(classification_report(y_test, y_pred_rf))
print("accuracy:", accuracy_score(y_test, y_pred_rf))
```

The Random Forest model slightly outperforms Logistic Regression in overall accuracy and precision, showing improved reliability when predicting confirmed exoplanets. This improvement comes at the cost of a small reduction in recall, meaning that some real planets are missed.


Comparison and Discussion

Both models perform well on the Kepler dataset. Logistic Regression is more sensitive
to detecting real planets, while Random Forest is more reserved and produces
fewer false positives. Overall, Random Forest provides the best balance between
precision and accuracy for this task.
